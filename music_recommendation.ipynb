```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib

music_data = pd.read_csv('./datasets/music.csv')
```

```python
X = music_data.drop(columns=['genre'])

y  = music_data['genre']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)            # allocating 20% of data for testing
```

```python
model = DecisionTreeClassifier()

# model.fit(X_train,y_train)          # we teach the model on the train data, and then use test data

model.fit(X,y)

plot_tree(model, feature_names = ['age', 'gender'],
                    class_names = sorted(y.unique()),
                     filled = True)

# predictions = model.predict(X_test)

# score = accuracy_score(y_test,predictions)

# score
```

### Joblib - dumping a model

```python
model = joblib.dump(model,'music-recommendation.joblib')
```
